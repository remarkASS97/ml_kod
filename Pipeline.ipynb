```python
#Загружаем библиотеки 
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
```

```python
#задаем начальное число для генератора псевдослучайных чисел
np.random.seed(0)
```

```python
#Загрузим данные 
iris = datasets.load_iris()
features = iris.data
target = iris.target
```

```python
#Создаем конвейер
pipe = Pipeline([('classifier', RandomForestClassifier())])
```

```python
#Создаем словарь вариантов обучающих алгоритмов и их гиперпараметров
search_space = [{'classifier': [LogisticRegression()],
                'classifier__penalty': ['l1', 'l2'],
                'classifier__C': np.logspace(0, 4, 10)},
               {'classifier': [RandomForestClassifier()],
                'classifier__n_estimators': [10, 100, 1000],
                'classifier__max_features': [1, 2, 3]}]
```

```python
#создаем объект решетчатого поиска
gridsearch = GridSearchCV(pipe, search_space, cv=5, verbose=0)
```

```python
#выполняем подгонку объекта решетчатого поиска
best_model = gridsearch.fit(features, target)
```

```python
# смотрим на лучшую модель
best_model.best_estimator_.get_params()['classifier']
```

```python
best_model.predict(features)
```
